```python
import pandas as pd
import numpy  as np
from tensorflow import keras
from tqdm import tqdm
import matplotlib.pyplot as plt
import random
from imblearn.over_sampling import SMOTE
import pydot
from tensorflow.keras.callbacks import EarlyStopping
import graphviz

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

def _train_test_split(df, test_prop):
    _sample_list = df[['gameId','playId']].drop_duplicates(keep='first').reset_index().drop('index',axis=1)
    for idx,row in _sample_list.iterrows():
        rand_val = random.random()
        if rand_val <= test_prop:
            _sample_list.at[idx,'TEST_FLAG'] = 1
        else:
            _sample_list.at[idx,'TEST_FLAG'] = 0


    test_plays = _sample_list[_sample_list['TEST_FLAG']==1].reset_index().drop('index',axis=1)
    train_plays = _sample_list[_sample_list['TEST_FLAG']==0].reset_index().drop('index',axis=1)

    pbar = tqdm(total=len(df))
    for idx,row in df.iterrows():
        if row['gameId'] in test_plays['gameId'].unique().tolist():
            if row['playId'] in test_plays[test_plays['gameId']==row['gameId']]['playId'].unique().tolist():
                df.at[idx,'TEST_FLAG'] = 1
            else:
                df.at[idx,'TEST_FLAG'] = 0
        else:
            df.at[idx,'TEST_FLAG'] = 0
        pbar.update(1)
    pbar.close()

    train = df[df['TEST_FLAG']==0].reset_index().drop('index',axis=1)
    test = df[df['TEST_FLAG']==1].reset_index().drop('index',axis=1)

    train = train.fillna(0)
    test = test.fillna(0)
    
    return train, test
```

```python
df = pd.read_csv('../data/processed/TRAINING_SAMPLE_ML_Dataset.csv').drop('Unnamed: 0',axis=1)
df = df[df['OFF_DEF_FLAG']=='DEF'].reset_index().drop('index',axis=1)
```

```python
df.head()
```

```python
# Dummying Categorical Fields

dummy_dict = {'offenseFormation':['SHOTGUN', 'EMPTY', 'I_FORM', 'SINGLEBACK', 'PISTOL', 'JUMBO', 'WILDCAT'],
              'position':['ILB', 'DT', 'CB', 'DE', 'FS', 'NT', 'SS', 'OLB', 'MLB', 'DB']}
    
for key,value in dummy_dict.items():
    
    _col = key
    _val_list = value
    
    for _val in _val_list:
        pbar = tqdm(total=len(df))
        for idx,row in df.iterrows():
            if row[_col] == _val:
                df.at[idx,f"{_col}_{_val}"] = 1
            else:
                df.at[idx,f"{_col}_{_val}"] = 0
            pbar.update(1)
        pbar.close()
```

```python
# Sampling For Train/Test Split Needs To Be By Game/PlayId
# (we don't want frames from 1 play being split across train/test)
(train, test) = _train_test_split(df, test_prop = 0.125)
```

```python
# Splitting into X & y Feature Sets (Info Set For Pred Analysis)
info_train = train[['gameId','playId','nflId','displayName','frameId','x','y','a','o','position']]
X_train = train[['x','y','s','a','dis','o','dir','weight','height','defendersInTheBox',
                 'YardsFromScoring','quarter','down','yardsToGo','Stadium_Domed','Stadium_Turfed','Kickoff_temp',
                 'Kickoff_dwpt','Kickoff_rhum','Kickoff_prcp','Kickoff_snow','Kickoff_wdir','Kickoff_wspd',
                 'Kickoff_pres','DISTANCE_TO_CLOSEST_OPP','DEF_ORIENTATION','ENGAGED_IN_BLOCK','DISTANCE_TO_BALL',
                 'OPEN_PATH_TO_BALL','FREE_BLOCKER_W_BALL_CARRIER_FLAG','offenseFormation_SHOTGUN','offenseFormation_EMPTY',
                 'offenseFormation_I_FORM','offenseFormation_SINGLEBACK','offenseFormation_PISTOL','offenseFormation_JUMBO',
                 'offenseFormation_WILDCAT','position_ILB','position_DT','position_CB','position_DE','position_FS',
                 'position_NT','position_SS','position_OLB','position_MLB','position_DB']]
y_train = train[target_col]

# OVERSAMPLING TO BALANCE TARGET FIELD
X_train_resampled, y_train_resampled = SMOTE().fit_resample(X_train, y_train)

info_test = test[['gameId','playId','nflId','displayName','frameId','x','y','a','o','position']]
X_test = test[['x','y','s','a','dis','o','dir','weight','height','defendersInTheBox',
             'YardsFromScoring','quarter','down','yardsToGo','Stadium_Domed','Stadium_Turfed','Kickoff_temp',
             'Kickoff_dwpt','Kickoff_rhum','Kickoff_prcp','Kickoff_snow','Kickoff_wdir','Kickoff_wspd',
             'Kickoff_pres','DISTANCE_TO_CLOSEST_OPP','DEF_ORIENTATION','ENGAGED_IN_BLOCK','DISTANCE_TO_BALL',
             'OPEN_PATH_TO_BALL','FREE_BLOCKER_W_BALL_CARRIER_FLAG','offenseFormation_SHOTGUN','offenseFormation_EMPTY',
             'offenseFormation_I_FORM','offenseFormation_SINGLEBACK','offenseFormation_PISTOL','offenseFormation_JUMBO',
             'offenseFormation_WILDCAT','position_ILB','position_DT','position_CB','position_DE','position_FS',
             'position_NT','position_SS','position_OLB','position_MLB','position_DB']]
y_test= test[target_col]
```

```python
len(X_train.columns.tolist())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = keras.Sequential(
    [
        keras.layers.Dense(units=27, activation="relu", input_shape=(X_train.shape[-1],) ),
#         # randomly delete 30% of the input units below
#         keras.layers.Dropout(0.3),
        keras.layers.Dense(units=15, activation="relu"),
        keras.layers.Dense(units=7, activation="relu"),
        # the output layer, with a single neuron
        keras.layers.Dense(units=1, activation="sigmoid"),
    ]
)

# save the initial weights for later
initial_weights = model.get_weights()
```

```python
model.summary()
```

```python
# keras.utils.plot_model(model, show_shapes=True)
```

```python
learning_rate = 0.001

model.compile(optimizer=keras.optimizers.Adam(learning_rate=learning_rate), 
              loss="binary_crossentropy", 
              metrics=keras.metrics.AUC()
             )
```

```python
history = model.fit(X_train, y_train, 
          epochs=5000, 
          batch_size=1000, 
          validation_data=(X_test, y_test),
          verbose=4)
```

```python
logs = pd.DataFrame(history.history)

plt.figure(figsize=(14, 4))
plt.subplot(1, 2, 1)
plt.plot(logs.loc[5:,"loss"], lw=2, label='training loss')
plt.plot(logs.loc[5:,"val_loss"], lw=2, label='validation loss')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(logs.loc[5:,"auc"], lw=2, label='training ROC AUC score')
plt.plot(logs.loc[5:,"val_auc"], lw=2, label='validation ROC AUC score')
plt.xlabel("Epoch")
plt.ylabel("ROC AUC")
plt.legend(loc='lower right')
plt.show()
```

```python
early_stopping = EarlyStopping(
    min_delta = 0.00001, # minimium amount of change to count as an improvement
    patience  = 20,     # how many epochs to wait before stopping
    restore_best_weights=True,
)
```

```python
model.set_weights(initial_weights)
```

```python
history = model.fit(X_train, y_train, 
          epochs=5000, 
          batch_size=1000, 
          validation_data=(X_test, y_test),
          verbose=4,
          # add in our early stopping callback
          callbacks=[early_stopping]
        )
```
