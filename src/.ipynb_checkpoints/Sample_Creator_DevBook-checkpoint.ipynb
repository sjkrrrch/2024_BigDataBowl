```python
import pandas as pd
import os
import numpy as np
import missingno as msno
```

```python
df = pd.read_csv('../data/raw/kaggle/plays.csv')
sample_df = df.sample(frac=0.15).reset_index().drop('index',axis=1)
sample_df = sample_df[['gameId','playId']]
sample_df.columns = ['SAMP_gameId','SAMP_playId']

training_df = pd.DataFrame()
for file in os.listdir('../data/processed'):
    if 'Week_' in file:
        df = pd.read_csv(f'../data/processed/{file}').drop('Unnamed: 0',axis=1)
        
        df = df.merge(sample_df,how='left',left_on=['gameId','playId'],right_on=['SAMP_gameId','SAMP_playId']).reset_index().drop('index',axis=1)
        df = df.dropna(subset=['SAMP_gameId','SAMP_playId']).reset_index().drop('index',axis=1)
        
        training_df = pd.concat([training_df,df]).reset_index().drop('index',axis=1)
```
